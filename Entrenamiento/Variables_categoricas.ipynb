Analyze a dataset by identifying numerical and categorical columns, calculating descriptive statistics for each type, and then providing a comprehensive summary and overview of the dataset.

## Cargar Dataset

### Subtask:
Cargar un dataset de ejemplo de una biblioteca de Python (por ejemplo, el dataset 'tips' de seaborn) en un DataFrame de pandas.


**Reasoning**:
First, I'll import the necessary libraries, pandas and seaborn, then load the 'tips' dataset into a pandas DataFrame named `df` as instructed.



```python
import pandas as pd
import seaborn as sns

# Load the 'tips' dataset from seaborn
df = sns.load_dataset('tips')

# Display the first few rows of the DataFrame to verify
print("Dataset 'tips' loaded successfully into DataFrame 'df'.")
df.head()
```

## Identificar Columnas

### Subtask:
Identificar y separar las columnas numéricas y categóricas del DataFrame cargado.


**Reasoning**:
To identify and separate numerical and categorical columns, I will use the `select_dtypes` method from pandas, first including numerical types, then including object and category types, and finally print the identified column names.



```python
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Columnas Numéricas: {numerical_cols}")
print(f"Columnas Categóricas: {categorical_cols}")
```

## Estadísticas Numéricas

### Subtask:
Calcular y mostrar las estadísticas descriptivas (como media, desviación estándar, mínimo, máximo, cuartiles) solo para las columnas numéricas.


**Reasoning**:
To calculate the descriptive statistics for the numerical columns, I will select these columns from the DataFrame `df` using the `numerical_cols` list and then apply the `.describe()` method.



```python
descriptive_stats_numerical = df[numerical_cols].describe()
print("Estadísticas Descriptivas para Columnas Numéricas:")
print(descriptive_stats_numerical)
```

## Estadísticas Categóricas

### Subtask:
Calcular y mostrar las estadísticas descriptivas (como conteo, número de valores únicos, valor más frecuente, frecuencia del valor más frecuente) solo para las columnas categóricas.


**Reasoning**:
To calculate the descriptive statistics for the categorical columns, I will select these columns from the DataFrame `df` using the `categorical_cols` list and then apply the `.describe()` method, printing the result with a descriptive header.



```python
descriptive_stats_categorical = df[categorical_cols].describe()
print("Estadísticas Descriptivas para Columnas Categóricas:")
descriptive_stats_categorical
```

## Estadísticas Completas

### Subtask:
Calcular y mostrar las estadísticas descriptivas para todas las columnas del dataset (numéricas y categóricas).


**Reasoning**:
To calculate descriptive statistics for all columns, including both numerical and categorical, I will apply the `.describe(include='all')` method directly to the DataFrame `df` and then print the result with a clear header.



```python
overall_descriptive_stats = df.describe(include='all')
print("Estadísticas Descriptivas para TODAS las Columnas:")
overall_descriptive_stats
```

```python
for col in categorical_cols:
    print(f"\n### Tabla de Frecuencias para '{col}':")
    absolute_freq = df[col].value_counts().rename('Frecuencia Absoluta')
    relative_freq = (df[col].value_counts(normalize=True) * 100).rename('Frecuencia Relativa (%)')

    # Combine into a single DataFrame for display
    frequency_table = pd.concat([absolute_freq, relative_freq], axis=1)
    display(frequency_table)
    print("-----------------------------------")
```

Existen las funciones.

Cuando hay que usar algo nuevamente, si estamos contruyendo un codigo mas grande, utilizo la función que se define con *def*

```python
def sumar(a,b):
  return a + b            #return se usa para resultado aritmético

def restar (a,b):
  return a - b

print(sumar(4,6))
print(restar(6,4))
```

```python
###promedio de ventas###
ventas =[120000, 85000, 43000, 3500, 100000]
print("Promedio:", sum(ventas) / len(ventas))
print("________________________")

#ventas superiores a 50000
for v in ventas:
  if v > 50000:
    print("menores a 50000: ", v)
print("________________________")

#cuente cuantas ventas se realizaron
print("Cantidad de ventas:", len(ventas))
print("_______________________")

#Aplicar IVA
valor = 100000
iva = valor*0.19
print("total con iva:", valor + iva)


```
